Задача 2. Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные
автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из
n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

 Сформулируем нулевую гипотезу:
$H0: \mu=\mu_0=17$,где $ \mu$ и $\mu_0$ - средние значения выборки и генеральной совокупности, соответственно.
Альтернативная гипотеза:
$H1: \mu>\mu_0$
Поскольку нам известны и среднее и дисперсия генеральной совокупности, то для проверки гипотезы будем использовать $z$ критерий.
$$z_н=\frac{\bar X-\mu_0}{\sigma/\sqrt n}$$,где:
$\bar X=\mu=17.5$ - среднее по выборке;
$\sigma=\sqrt 4=2$-  среднее квадратическое отклонение генеральной совокупности;
$n$=100 - размер выборки.

```python
z=(17.5-17)/0.2
print(f'Наблюдаемое значение z-критерия: {z: .4f}')
```

Найдем табличное значение  $z$-критерия, для $\alpha=0.05$

```python
import scipy.stats as stats
```

```python
zt=stats.norm.ppf(0.95)
print(f'Табличное значение z-критерия: {zt: .4f}')
```

Сравним полученные значения:

Поскольку $z_н>z_t$
, отвергаем нулевую гипотезу на уровне значимости $\alpha$

Ответ:Отвергаем нулевую гипотезу о равенствесредних значений на уровне значимости альфа.

Задача 3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья
составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна
99%? (Провести двусторонний тест.)

Из условия задачи следует что нужно сформулировать и протестировать следующие гипотезы:
$$H0: \mu=\mu_0=200$$,где $\mu$  и  $\mu_0$ средние выборки и генеральной совокупности, соответственно.

Альтернативная гипотеза:
$$H1: \mu\neq\mu_0$$
при этом будем использовать двусторонний критерий c уровнем значимости $\alpha=1-0.99=0.01$

Поскольку нам не известны ни среднее, ни дисперсия генеральной совокупности, то для проверки гипотезы будем использовать 
$t$-критерий.

```python
import numpy as np
```

```python
weight=np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
```

Расчитаем наблюдаемое значение t-критерия $t_н$:

```python
tn=(np.mean(weight)-200)/(np.std(weight,ddof=1)/len(weight)**0.5)
print(f'Наблюдаемое значение t-критерия: {tn: .4f}')
```

Найдем табличное значение t- критерия $t_t$, для $\alpha/2=0.01/2=0.005$

```python
tt=stats.t.ppf(0.995,len(weight)-1)
print(f'Табличное значение t-критерия: {tt: .4f}')
```

Поскольку $|t_н|<t_t$, принимаем нулевую гипотезу на уровне значимости $\alpha=0.01$

Задача 4. Есть ли статистически значимые различия в росте дочерей?

Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165

Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160, 163

Принимаем $alpha=0,05$
Используем двухвыборочную  $t$ проверку

```python
mothers=np.array([172,177,158,170,178,175,164,160,169,165])
daughters=np.array([173,175,162,174,175,168,155,170,160,163])
stats.ttest_rel (a=mothers, b=daughters)
```

$Pvalue$ > $alpha$ следовательно статистически значимых различий не обнаружено на уровне значимости $alpha$
